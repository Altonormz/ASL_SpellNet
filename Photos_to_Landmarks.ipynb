```python
import numpy as np 
import pandas as pd 
import cv2
import mediapipe as mp
import os
import gc
```

```python
PATH = '/media/altonormz-laptop/dok240/american-sign-language-09az/American'
```

```python
# We're using only hands in this notebook
# mp_holistic = mp.solutions.holistic
# holistic_model = mp_holistic.Holistic(
#     static_image_mode=True,
#     model_complexity=1,
#     min_detection_confidence=0.5,
#     min_tracking_confidence=0.5
# )

# Hand model
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()


def preprocess_image(img_path):
    # Read image and change from default BGR to RGB
    
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    if img is None:
        print('No picture was found at', img_path)
    return img

def process_image(img, holistic_model):
    # Return the results from the mediapipe model
    img.flags.writeable = False    
    results = holistic_model.process(img)
    return results
```

```python
def get_row_df(results, img_name):
    # Return a dataframe row with picture name and the landmarks of the hand.
    # Note this works only if there's one hand in picture.
    for hand in results.multi_hand_landmarks:
        x_landmarks = []
        y_landmarks = []
        z_landmarks = []
        for id, lm in enumerate(hand.landmark):
            x_landmarks.append(lm.x)
            y_landmarks.append(lm.y)
            z_landmarks.append(lm.z)

        landmarks = x_landmarks + y_landmarks + z_landmarks
        data = {'image_name': img_name}
        data.update({f'{dim}_left_{i}': landmarks[i] for dim in ['x', 'y', 'z'] for i in range(21)})
        row_df = pd.DataFrame([data])
    return row_df
```

```python
def get_row_dict(results, img_name):
    # Return a dictionary with picture name and the landmarks of the hand.
    # Note this works only if there's one hand in picture.
    for hand in results.multi_hand_landmarks:
        x_landmarks = []
        y_landmarks = []
        z_landmarks = []
        for id, lm in enumerate(hand.landmark):
            x_landmarks.append(lm.x)
            y_landmarks.append(lm.y)
            z_landmarks.append(lm.z)

        landmarks = x_landmarks + y_landmarks + z_landmarks
        data = {'image_name': img_name}
        data.update({f'{dim}_left_{i}': landmarks[i] for dim in ['x', 'y', 'z'] for i in range(21)})
    return data

total_folders = len(os.listdir(PATH))
for i,folder in enumerate(os.listdir(PATH)):
    
    print(f'Part {i}/{total_folders}')
    # Goes through every folder, and saves the folder name as label
    # Saves a csv file containing a subfolder landmarks from the photos inside.
    # Should save N csv files for N subfolder in main folder.
    rows_list = []
    label = folder
    folder_path = os.path.join(PATH,folder)
    for i,img_name in enumerate(os.listdir(folder_path)):
        img_path = os.path.join(folder_path,img_name)
        img_pr = preprocess_image(img_path)
        img_pr.flags.writeable = False
        results = hands.process(img_pr)
        if results.multi_hand_landmarks is not None:
            row_dict = get_row_dict(results, img_name)
            rows_list.append(row_dict)
        else:
            print('None')
    df = pd.DataFrame(rows_list)
    df.to_csv(f'dataframes/dataframe_for_{label}.csv',index=False)
    del df  # Trying to save ram
    gc.collect() 
print('Done')
```
